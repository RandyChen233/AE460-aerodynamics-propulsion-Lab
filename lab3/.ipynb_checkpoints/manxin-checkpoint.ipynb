```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

Processing raw data

# Task2

<font  size=2 > 
    Calculate and create a graph of the isentropic (or equivalent) Mach number using 
    &nbsp;   
<center>$M_{exit}=\sqrt{\frac{2}{\gamma -1}[(\frac{P_{0stag}}{P_{atm}}^{\frac{\gamma -1}{\gamma}})]}$<center/>

```python
gamma=1.4
M_sqaured = ((dfraw['Stagnation Pressure[Pa]']/dfraw['Absolute Static Pressure[Pa]'])**((gamma-1)/gamma) -1)/((gamma-1)/2)
M_isentropic = np.sqrt(M_sqaured)
M_isentropic
```

```python
P0_choked = 102731.92/ 0.52828178 #compressible flow calculator
P0_choked
```

```python
plt.figure(dpi=200);
#sns.lineplot(x='Stagnation Pressure[Pa]',y='M_isentropic',data=df,label='Isentropic Mach number')
plt.plot(p_stag,M_isentropic,'k',label='Assume isentropic')
plt.scatter(p_stag,M_isentropic,label='Data point')
plt.hlines(y=1.0,xmin=P0_choked,xmax=500000,color='k',linestyle='dotted',label='Choked flow adjustment')
plt.legend(fontsize=8)
plt.ylim(0.3,1.8);
plt.xticks(rotation=45);
plt.title('Isentropic Mach Number vs Stagnation Pressure[Pa]')
```

# Task 3

Calulate and create a graph for exit velocity determined from the stagnation chamber pressure from 3 different approaches:  
Bernoulli eqn(constant density):$V_{exit}=\sqrt{\frac{2(P_{0stag}-P_{atm})}{\rho}}$  
Assuming flow isentropic through out:$V_{exit}=M_{exit}\sqrt{\gamma RT_{amb}}$  
Assuming isentropic flow for subsonic isentropic Mach numbers and choked flow for supersonic insentropic Mach Number: $V_{exit}=\sqrt{\gamma RT_{exit}}$

#### approach 1 Bernoulli

```python
rho = 1.225 # kg/m^3 at STP,assume constant
V_exit_bernoulli = np.sqrt(2*(dfraw['Stagnation Pressure[Pa]']-dfraw['Absolute Static Pressure[Pa]'])/rho)
V_exit_bernoulli = V_exit_bernoulli.round(2)
V_exit_bernoulli
```

#### approach 2 Isentropic

```python
M_exit = np.sqrt(M_sqaured)
T0=295
T_exit = T0/(1+0.4/2*M_exit**2)
V_exit_isentropic = M_exit*np.sqrt(1.4*287*T_exit)
```

```python
V_exit_isentropic=V_exit_isentropic.round(2)
V_exit_isentropic
```

```python
T_exit_isentropic=T_exit.round(2)
T_exit_isentropic
```

#### approach 3 choke adjustment

```python
M_sqaured = ((dfraw['Stagnation Pressure[Pa]']/dfraw['Absolute Static Pressure[Pa]'])**((gamma-1)/gamma) -1)/((gamma-1)/2)

V_exit_adjusted = np.array([])
T_exit_adjusted = np.array([])
for i in dfraw['Stagnation Pressure[Pa]'].values:
    
    if i<102731.92/0.52828178:
        
        M_exit = np.sqrt(np.array(M_sqaured))
        T_exit = T0/(1+0.4/2*M_exit**2)
        T_exit_adjusted = np.append(T_exit_adjusted,T_exit)
        V_exit = np.array(M_exit*np.sqrt(1.4*287*T_exit))
        V_exit_adjusted = np.append(V_exit_adjusted,V_exit)
        
```

```python
V_exit_adjusted=np.unique(V_exit_adjusted[:3])
T_exit_adjusted=np.unique(T_exit_adjusted[:3])
T_exit_new = T0/1.2#/(1+0.4/2)
T_exit_adjusted_complete = np.append(T_exit_adjusted,T_exit_new*np.ones(5))
V_exit_new = np.array(np.sqrt(1.4*287*T_exit_new))
V_exit_adjusted_complete = np.append(V_exit_adjusted,V_exit_new*np.ones(5))
```

```python
V_exit_new
```

```python
V_choke_adjusted = V_exit_adjusted_complete.round(2)
V_choke_adjusted
```

```python
T_exit_new
```

```python
T_choke_adjusted = T_exit_adjusted_complete.round(2)
T_choke_adjusted 
```

```python
plt.figure(figsize=(10,9),dpi=100);
plt.plot(dfraw['Stagnation Pressure[Pa]'].values,V_exit_bernoulli,'--r',label='Exit velocity based on Bernoulli',linewidth=1.5)
plt.plot(dfraw['Stagnation Pressure[Pa]'].values,V_exit_isentropic,'-b',label='Isentropic exit velocity',linewidth=1.0,alpha=0.8)
plt.plot(dfraw['Stagnation Pressure[Pa]'].values,V_choke_adjusted,':g',label='Isentropic exit velocity',linewidth=1.5)
plt.legend()
plt.ylabel('Exit velocities[m/s]',fontsize=14)
plt.xlabel('Stagnation Pressure[Pa]',fontsize=14)
plt.ylim(100,850);
```

Calculate thrust from 3 different approach

```python
from math import pi
Jet_d = 0.2*0.0254 # convert to SI unit m
A_exit = pi*Jet_d**2/4 
```

```python
thrust_Bernoulli = V_exit_bernoulli**2*rho*A_exit
thrust_Bernoulli = thrust_Bernoulli.round(2)
thrust_Bernoulli
```

```python
thrust_isentropic = V_exit_isentropic**2*rho_isentropic*A_exit
thrust_isentropic = np.array(thrust_isentropic.round(2))
thrust_isentropic
```
