```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import find_peaks
```

Hydrogen

```python
hydrogen=pd.read_excel('lab5_data.xlsx', sheet_name='hydrogen')
hydrogen_back=pd.read_excel('lab5_data.xlsx', sheet_name='background_hydrogen')
hydrogen['emission']=hydrogen['Unnamed: 1']-hydrogen_back['Unnamed: 1']
```

```python
plt.plot(hydrogen['FLMT04185_1:101'], hydrogen['emission'])
plt.title('Hydrogen Emission Spectrum')
plt.xlabel('Wavelength (nm)')
plt.ylabel('Counts')
plt.ylim([0, 60000])
#plt.savefig('Hydrogen_Emission.png')
```

```python
Hpeaks, _ = find_peaks(hydrogen['emission'], height=850)
print(Hpeaks)
```
